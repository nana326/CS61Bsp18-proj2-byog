# Losses Data

```python
loss_file_path = '/content/Losses 1-1-2024 to 3-6-2024.xlsx'
```

```python
loss_data = pd.read_excel(loss_file_path, sheet_name = 'Export')
loss_data.head()
```

```python
# Eliminate hours and merge on shift start date
loss_data['event_start_date'] = pd.to_datetime(loss_data['eventStartTime']).dt.date
production_data_33['shift_start_date'] = pd.to_datetime(production_data_33['shiftstart_cst']).dt.date
loss_production_merged_full = pd.merge(loss_data, production_data_33, left_on='event_start_date', right_on='shift_start_date', how='right')
```

```python
loss_production_merged = loss_production_merged_full.drop(columns=['lineid', 'comment', 'updatedBy', 'event_start_date', 'shift_start_date'])
loss_production_merged.sort_values(by=['eventStartTime'])
```

```python
#rename columsn
merge_col_names= loss_production_merged.columns
merge_col_dict = {}

for i in merge_col_names:
  merge_col_dict[i] = i.lower().replace(" ", "_")

loss_production_merged = loss_production_merged.rename(columns = merge_col_dict)
loss_production_merged.head()

loss_production_merged = loss_production_merged.rename(columns = {"lossduration": "loss_duration_(min)",
                         "eventstarttime": "event_start_time",
                         "eventendtime": "event_end_time",
                          "faultvalue":"fault_value",
                          "causalmachine": "causal_machine",
                          "l1losstype": "l1_loss_type",
                          "l2losstypelocal":"l2_loss_type_local",
                         "l3losstypelocal":"l3_loss_type_local",
                         "l4losstypelocal":"l4_loss_type_local"})
loss_production_merged
```

```python
# group production data line 33 by dates and calculate the sum
production_data_33['date'] = pd.to_datetime(production_data_33['shiftstart_cst']).dt.date
line_33_production_date = production_data_33.groupby(['date']).sum().reset_index()
line_33_production_date
```

```python
#rename columns
loss_col_names= loss_data.columns
loss_col_dict = {}

for i in loss_col_names:
  loss_col_dict[i] = i.lower().replace(" ", "_")

loss_data = loss_data.rename(columns = loss_col_dict)
loss_data.head()
```

```python
remove_ucl = loss_data[loss_data["l1losstype"] != "UCL"]
remove_ucl
print(loss_data.shape)
print(remove_ucl.shape)

remove_ucl['date'] = pd.to_datetime(remove_ucl['eventstarttime']).dt.date
remove_ucl = remove_ucl.groupby(['date']).sum().reset_index()

remove_ucl.drop(columns = ["faultvalue"])

# remove_ucl.head()
```

```python
merged_by_day = pd.merge(remove_ucl, line_33_production_date, left_on='date', right_on='date', how='inner')
merged_by_day = merged_by_day.drop(columns = ["name"])
merged_by_day["loading_time"] = merged_by_day["loss_(h)"] + merged_by_day["vot_(h)"]
merged_by_day.head()

merged_by_day[merged_by_day["loading_time"]>24]
# print(remove_ucl.shape)
# print(line_33_production_date.shape)
# print(merged_by_day.shape)
```

```python
#add shifts
col = loss_data["eventstarttime"]
for j in range(len(col)):
    if col[j].hour >= 6 and col[j].hour < 18:
        loss_data.at[j,"shift"]= col[j].replace(hour=6, minute=0, second=0).strftime('%Y-%m-%d %H:%M:%S')
    elif col[j].hour >= 18:  # Hours between 6 PM and midnight
        loss_data.at[j,"shift"]= col[j].replace(hour=18, minute=0, second=0)
        loss_data.at[j,"shift"] = loss_data["shift"][j].strftime('%Y-%m-%d %H:%M:%S')
    else:  # Hours between midnight and 6 AM
        loss_data.at[j,"shift"] = col[j].replace(hour=18, minute=0, second=0) + timedelta(days=-1)
        loss_data.at[j,"shift"] = loss_data["shift"][j].strftime('%Y-%m-%d %H:%M:%S')
loss_data.head(20)
```

```python
#retry joining tables and creating loading time
loss_data['shift_date'] = None
for i in range(len(loss_data["shift"])):
  shift_datetime = datetime.strptime(loss_data["shift"][i], '%Y-%m-%d %H:%M:%S')
  loss_data.at[i, "shift_date"] = shift_datetime.date()

remove_ucl = loss_data[loss_data["l1losstype"] != "UCL"]
remove_ucl = remove_ucl.groupby(['shift_date']).sum().reset_index()

# remove_ucl.drop(columns = ["faultvalue"])

merged_by_day = pd.merge(remove_ucl, line_33_production_date, left_on='shift_date', right_on='date', how='inner')
merged_by_day = merged_by_day.drop(columns = ["name"])
merged_by_day["loading_time"] = merged_by_day["loss_(h)"] + merged_by_day["vot_(h)"]
merged_by_day.head()

merged_by_day
```
